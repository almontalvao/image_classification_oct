```python
import os 
base_dir = os.path.join('/content/drive/My Drive/fourthbrain/Capstone/OCT2017/')
resized_dir = os.path.join('/content/drive/My Drive/fourthbrain/Capstone/OCT2017_Resized/')
```

```python
size_per_class = 2500
size_per_class_labeled = size_per_class * 0.05
```

```python
def resize_images(base_dir, resized_dir):
    sub_directories = ['val'] # 'test', 'train' 
    class_names = ['CNV', 'DME', 'DRUSEN', 'NORMAL'] #  
    for sub_dir in sub_directories:
        for class_name in class_names:
            path_fragment = sub_dir + '/' + class_name
            input_dir = os.path.join(base_dir, path_fragment)
            output_dir = os.path.join(resized_dir, path_fragment)
            print('Creating directory', output_dir)
            try:
                os.makedirs(output_dir)
            except OSError as error:
                print(error)
            files = os.listdir(input_dir)
            i = 0
            for file_name in tqdm(files):
                if i >= size_per_class:
                  break
                input_image_path = input_dir + '/' + file_name
                image = cv2.imread(input_image_path)
                resized_image = cv2.resize(image, (150, 150), interpolation=cv2.INTER_AREA)
                output_image_path = output_dir + '/' + file_name
                cv2.imwrite(output_image_path, resized_image)
                i += 1
```

```python
resize_images(base_dir, resized_dir)
```

```python
labeled_dataset_size = 2664
size_per_class_labeled = labeled_dataset_size / 4
print('Labeled size per class:', size_per_class_labeled)
```

```python
def copy_images(base_dir):
    class_names = ['CNV', 'DME', 'DRUSEN', 'NORMAL']
    for class_name in class_names:
        input_dir = base_dir + 'train/' + class_name
        output_dir = base_dir + 'train_labeled/' + class_name
        print('Creating directory', output_dir)
        try:
            os.makedirs(output_dir)
        except OSError as error:
            print(error)
        files = os.listdir(input_dir)
        i = 0
        for file_name in tqdm(files):
            if i >= size_per_class_labeled:
              break
            input_image_path = input_dir + '/' + file_name
            image = cv2.imread(input_image_path)
            resized_image = cv2.resize(image, (150, 150), interpolation=cv2.INTER_AREA)
            output_image_path = output_dir + '/' + file_name
            cv2.imwrite(output_image_path, resized_image)
            i += 1
```

```python
copy_images(resized_dir)
```
